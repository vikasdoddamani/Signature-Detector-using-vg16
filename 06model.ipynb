```python
import numpy as np
import cv2
import glob
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint, LearningRateScheduler
from tensorflow.keras.applications import VGG16
import matplotlib.pyplot as plt
from sklearn.utils import shuffle
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.model_selection import KFold

# Define constants
IMAGE_SIZE = (224, 224)
BATCH_SIZE = 32
EPOCHS = 20

# Load and preprocess the data
def load_and_preprocess_image(image_path, label):
    image = tf.io.read_file(image_path)
    image = tf.image.decode_jpeg(image, channels=3)  # Assuming images are JPEG
    image = tf.image.resize(image, IMAGE_SIZE)  # Resize to 224x224
    image = image / 255.0  # Normalize to [0, 1]
    return image, label

# Load image paths and labels
gen = [
    glob.glob('./input/handwritten-signatures/Dataset_Signature_Final/Dataset/dataset1/real/*.*'),
    glob.glob('./input/handwritten-signatures/Dataset_Signature_Final/Dataset/dataset2/real/*.*'),
    glob.glob('./input/handwritten-signatures/Dataset_Signature_Final/Dataset/dataset3/real/*.*'),
    glob.glob('./input/handwritten-signatures/Dataset_Signature_Final/Dataset/dataset4/real1/*.*')
]

forg = [
    glob.glob('./input/handwritten-signatures/Dataset_Signature_Final/Dataset/dataset1/forge/*.*'),
    glob.glob('./input/handwritten-signatures/Dataset_Signature_Final/Dataset/dataset2/forge/*.*'),
    glob.glob('./input/handwritten-signatures/Dataset_Signature_Final/Dataset/dataset3/forge/*.*'),
    glob.glob('./input/handwritten-signatures/Dataset_Signature_Final/Dataset/dataset4/forge/*.*')
]

# Collect paths and labels
image_paths = []
labels = []

for data in range(len(gen)):
    for i in gen[data]:
        image_paths.append(i)
        labels.append(0)  # Genuine
    for j in forg[data]:
        image_paths.append(j)
        labels.append(1)  # Forged

# Shuffle and split the dataset (train/test split)
image_paths, labels = shuffle(image_paths, labels)

# Train-test split
train_paths = image_paths[:int(0.8 * len(image_paths))]
test_paths = image_paths[int(0.8 * len(image_paths)):]

train_labels = np.array(labels[:int(0.8 * len(labels))])
test_labels = np.array(labels[int(0.8 * len(labels)):])

# Create TensorFlow dataset
train_dataset = tf.data.Dataset.from_tensor_slices((train_paths, train_labels))
test_dataset = tf.data.Dataset.from_tensor_slices((test_paths, test_labels))

train_dataset = train_dataset.map(load_and_preprocess_image, num_parallel_calls=tf.data.AUTOTUNE)
train_dataset = train_dataset.shuffle(1000).batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)

test_dataset = test_dataset.map(load_and_preprocess_image, num_parallel_calls=tf.data.AUTOTUNE)
test_dataset = test_dataset.batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)

# Define the model with fine-tuning
def build_model():
    base_model = VGG16(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
    base_model.trainable = False  # Freeze the convolutional base initially

    # Add custom layers on top of the pre-trained base model
    x = layers.Flatten()(base_model.output)
    x = layers.Dense(128, activation='relu')(x)
    x = layers.BatchNormalization()(x)  # Batch normalization for better training stability
    x = layers.Dropout(0.3)(x)
    x = layers.Dense(2, activation='softmax')(x)  # Output layer: 2 classes (Genuine, Forged)

    model = models.Model(inputs=base_model.input, outputs=x)
    model.compile(optimizer=Adam(learning_rate=0.001), loss="sparse_categorical_crossentropy", metrics=["accuracy"])
    return model

# Build and compile the model
network = build_model()
network.summary()

# Callbacks
earlyStopping = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True, verbose=1)

# Learning rate scheduler to reduce LR when validation loss plateaus
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=2, min_lr=1e-6, verbose=1)

# Save the best model
checkpoint = ModelCheckpoint('best_signature_model.keras', monitor='val_loss', save_best_only=True, verbose=1)

# Define cyclical learning rate scheduler
def cyclical_learning_rate(epoch):
    max_lr = 0.001
    base_lr = 0.0001
    step_size = 5  # Period of one cycle
    cycle = np.floor(1 + epoch / (2 * step_size))
    x = np.abs(epoch / step_size - 2 * cycle + 1)
    lr = base_lr + (max_lr - base_lr) * np.maximum(0, (1 - x))
    return lr

clr_scheduler = LearningRateScheduler(cyclical_learning_rate)

# Train the model
class_weight = {0: 1., 1: 3.}  # Example, adjust based on your dataset's balance
network.fit(train_dataset, epochs=EPOCHS, validation_data=test_dataset, class_weight=class_weight, callbacks=[earlyStopping, reduce_lr, checkpoint, clr_scheduler])

history = network.fit(
    train_dataset, 
    epochs=EPOCHS, 
    validation_data=test_dataset, 
    class_weight=class_weight, 
    callbacks=[earlyStopping, reduce_lr, checkpoint, clr_scheduler]
)


# Plot training and validation accuracy and loss
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']

epochs = range(len(acc))

# Plot Training & Validation Accuracy
plt.plot(epochs, acc, 'b', label='Training acc')
plt.plot(epochs, val_acc, 'r', label='Validation acc')
plt.title('Training and Validation Accuracy')
plt.legend()

# Plot Training & Validation Loss
plt.figure()
plt.plot(epochs, loss, 'b', label='Training loss')
plt.plot(epochs, val_loss, 'r', label='Validation loss')
plt.title('Training and Validation Loss')
plt.legend()

plt.show()

# Save the model
network.save('signature_model.keras')

# Evaluate model performance with more metrics
# def evaluate_model():
#     predictions = network.predict(test_paths)
#     pred_labels = np.argmax(predictions, axis=1)
    
#     print("Classification Report:")
#     print(classification_report(test_labels, pred_labels))
    
#     print("Confusion Matrix:")
#     print(confusion_matrix(test_labels, pred_labels))

# # Evaluate the model
# evaluate_model()

def evaluate_model():
    # Preprocess the test images
    def preprocess_image(image_path):
        # Load and preprocess each image from the path
        image = tf.io.read_file(image_path)
        # image = tf.image.decode_png(image, channels=3)  # Assuming images are png
        try:
            image = tf.image.decode_image(image, channels=3)  # Decode the image to RGB
        except Exception as e:
            print(f"Error decoding image {image_path}: {e}")
            return None
        image = tf.image.resize(image, IMAGE_SIZE)  # Resize to 224x224
        image = image / 255.0  # Normalize to [0, 1]
        return image
    
    # Apply preprocessing to each image in the test set
    test_images = [preprocess_image(path) for path in test_paths]
    test_images = np.array(test_images)
    
    # Make predictions
    predictions = network.predict(test_images)
    pred_labels = np.argmax(predictions, axis=1)
    
    # Print classification report and confusion matrix
    print("Classification Report:")
    print(classification_report(test_labels, pred_labels))
    
    print("Confusion Matrix:")
    print(confusion_matrix(test_labels, pred_labels))

# Evaluate the model
evaluate_model()


# Example of how to test with a new image
def predict_image(image_path):
    # Load the image
    image = cv2.imread(image_path)
    
    # Check if the image was loaded correctly
    if image is None:
        print("Error: Image not found or path is incorrect.")
        return
    
    # Convert from BGR to RGB (OpenCV loads images in BGR format)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    
    # Resize the image to match the model's input size
    image = cv2.resize(image, IMAGE_SIZE)  # 224x224 size
    
    # Normalize the image (scale pixel values between 0 and 1)
    image = np.array(image) / 255.0
    
    # Expand dimensions to match the input shape of the model (batch size of 1)
    image = np.expand_dims(image, axis=0)  # Shape becomes (1, 224, 224, 3)

    # Make predictions
    predictions = network.predict(image)
    
    # Get the predicted class (0 for Genuine, 1 for Forged)
    class_idx = np.argmax(predictions, axis=1)
    
    if class_idx == 0:
        print("Predicted Class: Genuine Signature")
    else:
        print("Predicted Class: Forged Signature")

# Test with a new image (use a valid image path)
image_path = './real.png'  # Replace with your image path
predict_image(image_path)

```
